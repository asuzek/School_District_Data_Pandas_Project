## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
school_data_comp_math=school_data_complete.drop(["Student ID", "reading_score", "School ID", "size", "budget"], axis=1)
school_data_math=school_data_comp_math.groupby([ "grade", "school_name"]).mean()
df_m=school_data_math
#print(df_m)
df_m.reset_index(level=df_m.index.names, inplace=True)
#print(df_m)
df_m_9=df_m.loc[df_m["grade"]=="9th",:]
df_m_10=df_m.loc[df_m["grade"]=="10th",:]
df_m_11=df_m.loc[df_m["grade"]=="11th",:]
df_m_12=df_m.loc[df_m["grade"]=="12th",:]
df_m_9_10=pd.merge(df_m_9, df_m_10, on="school_name", how="inner")
df_m_11_12=pd.merge(df_m_11, df_m_12, on="school_name", how="inner")
df_m_merged=pd.merge(df_m_9_10, df_m_11_12, on="school_name", how="inner")
#print(df_m_merged)
df_merged=df_m_merged.drop(['grade_x_x', 'grade_y_x', 'grade_x_y', 'grade_y_y'], axis=1)
df_m_fin=df_merged.rename(columns={"math_score_x_x":"9th", "math_score_y_x":"10th", "math_score_x_y":"11th", "math_score_y_y":"12th"})
df_m_fin.set_index(["school_name"], inplace=True)
df_m_fin.index.names = [""]
df_m_fin
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
school_data_comp_read=school_data_complete.drop(["Student ID", "math_score", "School ID", "size", "budget"], axis=1)
school_data_read=school_data_comp_read.groupby([ "grade", "school_name"]).mean()
df_r=school_data_read

df_r.reset_index(level=df_r.index.names, inplace=True)

df_r_9=df_r.loc[df_r["grade"]=="9th",:]
df_r_10=df_r.loc[df_r["grade"]=="10th",:]
df_r_11=df_r.loc[df_r["grade"]=="11th",:]
df_r_12=df_r.loc[df_r["grade"]=="12th",:]
df_r_9_10=pd.merge(df_r_9, df_r_10, on="school_name", how="inner")
df_r_11_12=pd.merge(df_r_11, df_r_12, on="school_name", how="inner")
df_r_merged=pd.merge(df_r_9_10, df_r_11_12, on="school_name", how="inner")
df_merged=df_r_merged.drop(['grade_x_x', 'grade_y_x', 'grade_x_y', 'grade_y_y'], axis=1)
df_r_fin=df_merged.rename(columns={"reading_score_x_x":"9th", "reading_score_y_x":"10th", "reading_score_x_y":"11th", "reading_score_y_y":"12th"})
df_r_fin.set_index(["school_name"], inplace=True)
df_r_fin.index.names = [""]
df_r_fin
```
